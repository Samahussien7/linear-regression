```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
x = np.arange(200)
delta = np.random.uniform(-50,30, size=(200,))
y = .4 * x +3 + delta

plt.scatter(x, y)
plt.show()
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x, y, random_state=0, train_size = .80)
```

```python
def costfun(x,y,w,b):
    total=0.0
    y_p=w*x+b
    total=0.5*(np.mean((y_p-y)**2))
    return total
```

```python
m=np.random.uniform(low=0.0, high=1.0, size=None)
c=np.random.uniform(low=0.0, high=1.0, size=None)

L = 0.000001  
epochs = 100000 

n = float(len(xtrain)) 
for i in range(epochs): 
    Y_pred = m*xtrain + c  
    cost=.5*(np.mean((ytrain - Y_pred)**2))
    D_m = (1/n) * sum(xtrain * ( Y_pred-ytrain ))  
    D_c = (1/n) * sum(Y_pred-ytrain )  
    m = m - L* D_m  
    c = c - L* D_c  
    
print (m, c,cost)
```

```python

costfun(xtest,ytest,m,c)
```

```python
plt.scatter(x, y)
yf=m*xtest+c
plt.plot(xtest,yf,color="black")
plt.show()
```
